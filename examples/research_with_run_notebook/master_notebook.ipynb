```python
import os
import numpy as np
import matplotlib.pyplot as plt
import time

time.sleep(np.random.uniform(5))
```

```python
DISTRIBUTION = 'normal'
SIZE = 1000
DST = '.'
```

```python
array = getattr(np.random, DISTRIBUTION)(size=SIZE)
mean, std = array.mean(), array.std()
```

```python
vals, _, _ = plt.hist(array, bins=SIZE // 10, density=True)
plt.vlines(mean, 0, vals.max(), colors='orange')
plt.suptitle(f'{DISTRIBUTION} distribution (size {SIZE})')
plt.title(f"mean={mean:.03f}   std={std:.03f}")
plt.savefig(os.path.join(DST, 'hist.png'))
```
